```python
# Storing Data

lokasi_list = []
luas_tanah_list = []
luas_bangunan_list = []
carport_list = []
kamar_tidur_list = []
kamar_mandi_list = []
harga_list = []

total_pages = 10

link_url = "https://www.rumah123.com/jual/bekasi/rumah/?page="
```

### Insights : Above loop (1st looping) is needed to test the container and it has successfully been done. 

```python
for page in range(1, total_pages + 1):

    print("Scraping page:", page)

    page_url = link_url + str(page)
    driver.get(page_url)
    time.sleep(5)

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    house_list = soup.find_all("div", {"data-name": "ldp-listing-content"})

    for house in house_list:

        # =========================
        # LOKASI
        # =========================
        lokasi_element = house.find("p")
        if lokasi_element:
            lokasi_list.append(lokasi_element.text.strip())
        else:
            lokasi_list.append(None)

        # =========================
        # DETAIL INFO (KT, KM, Carport, LT, LB)
        # =========================
        info_box = house.find("div", class_="bg-lightBlue")

        kamar_tidur = None
        kamar_mandi = None
        carport = None
        luas_tanah = None
        luas_bangunan = None

        if info_box:

            span_items = info_box.find_all("span")

            for item in span_items:
                text = item.text.strip()

                # Bedroom
                if "+" in text and kamar_tidur is None:
                    kamar_tidur = text

                # Bathroom
                elif "+" in text and kamar_tidur is not None:
                    kamar_mandi = text

                # Carport (only number)
                elif text.isdigit():
                    carport = text

                # Luas Tanah & Luas Bangunan (both contain m²)
                elif "m²" in text:
                    if luas_tanah is None:
                        luas_tanah = text.replace("LT:", "").strip()
                    else:
                        luas_bangunan = text.replace("LB:", "").strip()

        kamar_tidur_list.append(kamar_tidur)
        kamar_mandi_list.append(kamar_mandi)
        carport_list.append(carport)
        luas_tanah_list.append(luas_tanah)
        luas_bangunan_list.append(luas_bangunan)

        # =========================
        # HARGA
        # =========================
        harga_element = house.find("span", {"data-testid": "ldp-text-price"})
        if harga_element:
            harga_list.append(harga_element.text.strip())
        else:
            harga_list.append(None)

print("Scraping DONE")
```

## DATA CLEANING 

- Handling Missing Value bila ada.
- Mengubah bentuk data agar menjadi lebih konsisten.
- Menyesuaikan tipe data.
- Perlu diingat bahwa data ini akan digunakan untuk perhitungan statisik dan membutuhkan kolom dengan tipe data angka.

```python
# CLEAN : HARGA RUMAH

def clean_harga(price):
    if pd.isna(price):
        return None
    
    # Ambil hanya harga pertama (sebelum Rp kedua / promo)
    first_price = price.split("Rp")[1]
    first_price = first_price.split("Rp")[0].strip()
    
    # Hilangkan kata promo jika ada
    first_price = first_price.split("HEMAT")[0].strip()
    
    if "Miliar" in first_price:
        value = float(first_price.replace("Miliar", "").replace(",", ".").strip())
        return value * 1000  # ubah ke jutaan
    
    elif "Juta" in first_price:
        value = float(first_price.replace("Juta", "").replace(",", ".").strip())
        return value
    
    return None

data["Harga Rumah"] = data["Harga Rumah"].apply(clean_harga)
```

```python
# CLEAN : LOKASI

data["Lokasi"] = data["Lokasi"].str.replace(", Bekasi", "")
```

```python
# CARPORT : MISSING > 0 

data["Carport"] = data["Carport"].fillna(0)
data["Carport"] = data["Carport"].astype(int)
```

```python
# CLEAN : JUMLAH KAMAR TIDUR

def clean_kamar(value):
    if pd.isna(value):
        return 0
    
    value = str(value)
    
    if "+" in value:
        parts = value.split("+")
        return int(parts[0]) + int(parts[1])
    else:
        return int(value)

data["Jumlah Kamar Tidur"] = data["Jumlah Kamar Tidur"].apply(clean_kamar)
data["Jumlah Kamar Mandi"] = data["Jumlah Kamar Mandi"].apply(clean_kamar)
```

```python
# CLEAN : LUAS TANAH & LUAS BANGUNAN

data["Luas Tanah"] = data["Luas Tanah"].astype(float)
data["Luas Bangunan"] = data["Luas Bangunan"].astype(float)
```

### 2. Anda perlu mempersiapkan uang untuk pembelian rumah ini sehingga Anda harus mengetahui berapa kisaran harga rumah yang wajar di Bekasi ?

*Gunakan confidence interval untuk mendapatkan lower value dan upper value dari distribusi populasi harga rumah. Gunakan confidence level 95%.

```python
harga = data["Harga Rumah"]

mean = harga.mean()
std = harga.std()
n = len(harga)

# Standard Error
se = std / (n ** 0.5)

# Confidence Interval 95%
lower = mean - 1.96 * se
upper = mean + 1.96 * se

print("Mean Harga Rumah:", mean)
print("Lower Bound (95%):", lower)
print("Upper Bound (95%):", upper)
```

### Insights

Dengan menggunakan confidence level 95%, rata-rata harga rumah di Bekasi diperkirakan berada di antara 1404 juta (lower bound) hingga 2020 juta (upper bound). Selain itu, nilai rata-rata harga rumah adalah sekitar Rp 1.7 Miliar. 

### Kesimpulan : 
Ini berarti, kisaran harga rumah yang wajar di Bekasi berada di sekitar Rp 1,4 Miliar - Rp 2 Miliar
- Jadi, pelanggan bisa menargetkan harga rumah di Rp 1.7 Miliar untuk optimalisasi harga yang terbaik.


### 3. Saat ini Anda belum memiliki mobil sehingga Carport bukan prioritas saat ini. Anda mau tahu apakah rumah dengan Carport lebih mahal dibandingkan tanpa Carport. Berikan analisa apakah layak membayar lebih untuk rumah yang memiliki Carport? 

Gunakan uji hipotesis yang diawali dengan menuliskan hipotesis null dan alternatifnya serta tentukan jenis hipotesis yang digunakan.

```python
from scipy.stats import ttest_ind

# Pisahkan harga berdasarkan carport
harga_carport = data[data["Carport"] > 0]["Harga Rumah"]
harga_tanpa_carport = data[data["Carport"] == 0]["Harga Rumah"]

# Independent t-test
t_stat, p_value_two_tailed = ttest_ind(harga_carport, harga_tanpa_carport)

# Because this is one-tailed test:
p_value_one_tailed = p_value_two_tailed / 2

print("T-statistic:", t_stat)
print("P-value (two-tailed):", p_value_two_tailed)
print("P-value (one-tailed):", p_value_one_tailed)
```

## Insight : 

Hasil : 
T-statistic = -7.17
P-value (one-tailed) ≈ 7.46e-12 

P-value sangat kecil, jauh < 0.05, Tapi t-statistic NEGATIF

Maka : 
Rata-rata harga rumah dengan carport lebih rendah daripada tanpa carport. 

## Kesimpulan : 

Berdasarkan hasil analisa diatas, tidak ada bukti bahwa rumah dengan carport lebih mahal. Bahkan, rata-ratanya cenderung lebih rendah. Jadi secara statistik, tidak ada indikasi bahwa carport membuat harga rumah lebih tinggi. Dengan demikian, pelanggan tidak layak untuk membayar lebih mahal dengan adanya carport.

### 4 . Jika harus memilih, apakah Anda lebih baik membeli tanah yang luas ataukah membeli bangunan yang luas ? Manakah yang lebih mempengaruhi harga rumah, apakah Luas Tanah ataukah Luas Bangunan ? Berikan pendapatmu.

Anda dapat jawab pertanyaan ini dengan uji korelasi. Gunakan library SciPy (bukan Pandas). Analisis nilai korelasi dan p-value nya. Gunakan teknik yang sesuai dengan kondisi data !

```python
from scipy.stats import pearsonr

# Luas Tanah & Harga
corr_tanah, p_tanah = pearsonr(data["Luas Tanah"], data["Harga Rumah"])

# Luas Bangunan & Harga
corr_bangunan, p_bangunan = pearsonr(data["Luas Bangunan"], data["Harga Rumah"])

print("Korelasi Luas Tanah vs Harga:", corr_tanah)
print("P-value Luas Tanah:", p_tanah)

print("Korelasi Luas Bangunan vs Harga:", corr_bangunan)
print("P-value Luas Bangunan:", p_bangunan)
```

### Insight 

Kedua korelasi cukup tinggi, artinya : 
- semakin luas tanah, harga rumah semakin tinggi.
- semakin luas bangunan, harga rumah juga meningkat.

Kita bisa memakai dari nilai korelasi : 
- Luas Tanah = 0.894
- Luas Bangunan = 0.844

--> luas tanah sedikit lebih tinggi dalam menentukan harga.

### Kesimpulan 

Berdasarkan hasil uji korelasi Pearson, baik Luas Tanah maupun Luas Bangunan sama-sama punya hubungan positif yang sangat kuat dengan harga rumah, dan hasilnya juga signifikan secara statistik (p-value < 0.05). Tapi kalau dibandingkan, korelasi Luas Tanah (0.894) sedikit lebih tinggi daripada Luas Bangunan (0.844).
